```python
#imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

```python
#Caso de teste 2 arch 1

print('Caso de teste 2 arch 1')
arquivo = np.load('teste2.npy')
x = arquivo[0]
y = np.ravel(arquivo[1])

hidden_layer_size = 2
max_iter = 10000
n_iter_no_change = 500
activation = 'relu'

regr = MLPRegressor(hidden_layer_sizes=(hidden_layer_size),
                    max_iter=max_iter,
                    activation=activation, #{'identity', 'logistic', 'tanh', 'relu'},
                    solver='adam',
                    learning_rate = 'adaptive',
                    n_iter_no_change=n_iter_no_change)
print('\nTreinando RNA com:')
print('Número de camadas escondidas: ', hidden_layer_size)
print('Número de iterações: ', max_iter)
print('Número de iterações sem mudanças:', n_iter_no_change)
print('Ativação: ', activation)

regr = regr.fit(x,y)



print('\nPreditor')
y_est = regr.predict(x)

#plot do primeiro caso de teste

plt.figure(figsize=[14,7])

#plot curso original
plt.subplot(1,3,1)
plt.plot(x,y)

#plot aprendizagem
plt.subplot(1,3,2)
plt.plot(regr.loss_curve_)

#plot regressor
plt.subplot(1,3,3)
plt.plot(x,y,linewidth=1,color='yellow')
plt.plot(x,y_est,linewidth=2)




plt.show()
```
